### Load the dataset

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('car_data.csv')
df
```

### Create a histogram of Car Prices 

```python
plt.hist(df.price, bins=40, alpha=0.9, color='steelblue', edgecolor='black')

# Histrogram Labels 
plt.title('Histogram of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```
