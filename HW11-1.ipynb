### Load the dataset

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('cities.csv')
df
```

### Create a Scatter Plot  

```python
x = df.longd
y = df.latd

colors = np.log10(df.population_total)
sizes = df.area_total_sq_mi

plt.scatter(x, y, 
            c=colors, s=sizes,
            alpha=0.8,
            cmap='turbo')

plt.clim(3.0,7.0)
cb = plt.colorbar()

#Scatter Plot labels 
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Cities: Area and Population')
SUB = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
log = "log10(population)".translate(SUB)
cb.set_label(log)
```
